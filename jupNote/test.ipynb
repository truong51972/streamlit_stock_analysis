```python
from vnstock import *
import pandas as dp
from datetime import date, timedelta
```

```python
last_quaters = 50
incomestatement = financial_flow(symbol="FPT", report_type='incomestatement', report_range='quarterly')
incomestatement.reset_index(inplace=True)
incomestatement = incomestatement.loc[:last_quaters] 

balancesheet = financial_flow(symbol="FPT", report_type='balancesheet', report_range='quarterly')
balancesheet.reset_index(inplace=True)
balancesheet = balancesheet.loc[:last_quaters] 

cashflow = financial_flow(symbol="FPT", report_type='cashflow', report_range='quarterly')
cashflow.reset_index(inplace=True)
cashflow = cashflow.loc[:last_quaters] 

financialRatio = financial_ratio('FPT', 'quarterly', is_all=True).T
financialRatio.reset_index(inplace=True)
financialRatio = financialRatio.loc[:last_quaters] 

# financial_ratio('FPT', 'quarterly', is_all=True).T['roe'].values
```

```python
financial_df = pd.DataFrame()
financial_df['Quý'] = incomestatement['index']
financial_df['Tỷ suất lợi nhuận gộp (GPM)'] = round(incomestatement['grossProfit']/incomestatement['revenue'], 3)
financial_df['Tỷ suất lợi hoạt động (OPM)'] = round(incomestatement['operationProfit']/incomestatement['revenue'], 3)
financial_df['Tỷ suất trước thuế (PTM)'] = round(incomestatement['preTaxProfit']/incomestatement['revenue'], 3)
financial_df['Tỷ suất sau thuế (NPM)'] = round(incomestatement['postTaxProfit']/incomestatement['revenue'], 3)
financial_df['Tỷ suất sinh lời trên vốn chủ (ROE)'] = financialRatio['roe']
financial_df['Tỷ suất sinh lời trên tài sản (ROA)'] = financialRatio['roa']
# financial_df['Tỷ suất sinh lời trên vốn đầu tư (ROIC)'] = round(incomestatement['postTaxProfit']/balancesheet['asset'], 3)
financial_df
```
